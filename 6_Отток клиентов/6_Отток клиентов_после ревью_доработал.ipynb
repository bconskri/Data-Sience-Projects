## 2.1 Обучим модели, оценим метрики

```python
#обучим модели
def metrics_calc(model, target_valid, prediction, features_valid):
    print('Матрица ошибок:')
    display(confusion_matrix(target_valid, prediction))
    print("Полнота" , recall_score(target_valid, prediction))
    print("Точность", precision_score(target_valid, prediction))
    print("F1-мера", f1_score(target_valid, prediction, labels=np.unique(prediction)))
    probabilities_valid = model.predict_proba(features_valid)[:, 1]
    print("AUC-ROC", roc_auc_score(target_valid, probabilities_valid))
    #построим график ROC кривой, чтобы увидеть насколько предсказание модели близко к случайной=константной
    fpr, tpr, thresholds = roc_curve(target_valid, probabilities_valid) 

    plt.figure()
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
    
    return
    
def models_train_score(p_features_train, p_target_train, p_features_valid, p_target_valid, p_class_weight = None):
    #2.1 дерево решений
    best_result = 0
    predictions_on_best_model = None
    for depth in range(1,22):
        model = DecisionTreeClassifier(random_state=12345, max_depth=depth, class_weight = p_class_weight)
        model.fit(p_features_train, p_target_train)
        predictions_valid  = model.predict(p_features_valid)
        result = accuracy_score(p_target_valid, predictions_valid)
        if result > best_result:
            model_DTC = model
            best_result = result
            predictions_on_best_model = predictions_valid

    print("Accuracy наилучшей модели DTC на валидационной выборке:", best_result)
    print('Глубина наилучшего дерева:', model_DTC.get_depth())
    metrics_calc(model_DTC, target_valid, predictions_on_best_model, features_valid)
    print()

    #2.2 случайный лес
    best_result = 0
    best_est = 0
    predictions_on_best_model = None
    for est in range(1, 100):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, class_weight = p_class_weight) # обучите модель с заданным количеством деревьев
        model.fit(p_features_train, p_target_train) # обучите модель на тренировочной выборке
        predictions_valid  = model.predict(p_features_valid)
        result = accuracy_score(p_target_valid, predictions_valid)
        if result > best_result:
            model_RFC = model# сохраните наилучшую модель
            best_result = result#  сохраните наилучшее значение метрики accuracy на валидационных данных
            best_est = est
            predictions_on_best_model = predictions_valid

    print("Accuracy наилучшей модели RFC на валидационной выборке:", best_result)
    print('Estimators наилучшего дерева:', best_est)
    print('Глубина наилучшего дерева:', model_RFC.get_params()['max_depth'])
    print()
    metrics_calc(model_RFC, target_valid, predictions_on_best_model, features_valid)
    print()

    #2.3 логистическая регрессия
    model_LR = LogisticRegression(random_state=12345, solver = 'liblinear', class_weight = p_class_weight) # инициализируйте модель логистической регрессии с параметром random_state=12345
    model_LR.fit(p_features_train, p_target_train) # обучите модель на тренировочной выборке
    predictions_valid  = model_LR.predict(p_features_valid)
    result = accuracy_score(p_target_valid, predictions_valid)

    print("Accuracy модели LR на валидационной выборке:", result)
    print()
    metrics_calc(model_LR, target_valid, predictions_valid, features_valid)
    print()
    
    return model_DTC, model_RFC, model_LR
    
#обучим и оценим модели
model_DTC, model_RFC, model_LR = models_train_score(features_train, target_train, features_valid, target_valid)
```

# 3. Борьба с дисбалансом

В обучающией и валидационной выборок заметный дисбаланс класов.
0 - 80%
1 - 20%

для тогно чтобы сбалансировать выборку необходимо в 4 раза увеличить количество положительных ответов.  
Для этого:
- разделим обучающую выборку на отрицательные и положительные объекты  
- скопируем несколько раз положительные объекты  
- с учётом полученных данных создадим новую обучающую выборку  
- перемешаем данные  
- обучим новые модели  


## 3.1 Сбалансируем выборку

```python
#функция увеличения выборки
def upsample(features, target, repeat, upsampled_сlass):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    if upsampled_сlass == 0:
        features_upsampled = pd.concat([features_zeros]* repeat + [features_ones] )
        target_upsampled = pd.concat([target_zeros]* repeat + [target_ones] )
        features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
        
    elif upsampled_сlass == 1:
        features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
        target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
        features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

#увеличим положительные ответы
features_train_upsampled, target_train_upsampled = upsample(features_train, target_train, 4, 1)
display(target_train_upsampled.value_counts(normalize = 1))
display(target_train_upsampled.shape)
```

Классы сбаланисровали. Можно обучить модель на новых данных
