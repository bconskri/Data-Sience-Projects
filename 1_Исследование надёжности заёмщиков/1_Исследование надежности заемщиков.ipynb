```python
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∑–∞–ø–∏—Å–∏ –≤ –∫–æ—Ç–æ—Ä—ã—Ö –≤ –∑–Ω–∞—á–∏–º—ã—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö –ø—É—Å—Ç–æ
print('–ü—Ä–æ–ø—É—Å–∫ –¥–∞–Ω–Ω—ã—Ö')
print('–î–µ—Ç–∏ NA', data['children'].isna().sum())
print('–°–µ–º–µ–π–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ NA', data['family_status'].isna().sum())
print('–£—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ NA', data['total_income'].isna().sum())
print('–¶–µ–ª–∏', data['purpose'].isna().sum())
print('–£—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ –ø—Ä–æ–ø—É—â–µ–Ω, –Ω–æ –µ—Å—Ç—å –¥–æ–ª–≥–∏', len(data[data['total_income'].isna() & data['debt'] == 1]))
print('–£—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ –ø—Ä–æ–ø—É—â–µ–Ω, –Ω–æ –µ—Å—Ç—å –¥–æ–ª–≥–∏', data[data['total_income'].isna() & data['debt'] == 1])
print('–£—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ 0, –Ω–æ –µ—Å—Ç—å –¥–æ–ª–≥–∏', data[(data['total_income'] == 0) & (data['debt'] == 1)])
print()
#–∑–∞–º–µ–Ω–∏–º –ø—É—Å—Ç–æ–π –¥–æ—Ö–æ–¥ –Ω–∞ –º–µ–¥–∏–∞–Ω—É
#–ø—Ä–∏ —ç—Ç–æ–º –ø–æ—Å—á–∏—Ç–∞–µ–º –º–µ–¥–∏–∞–Ω—É –ø–æ –∫–∞–∂–¥–æ–º—É –≤–∏–¥—É –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ –æ—Ç–¥–µ–ª—å–Ω–æ
# —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—á–∫—É —Å –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–æ–π –ø–æ —Ç–∏–ø—É –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ –∏ –∏–∑–º–µ—Ä–µ–Ω–∏–µ–º –º–µ–¥–∏–∞–Ω–Ω–æ–≥–æ –¥–æ—Ö–æ–¥–∞
group_income = data.groupby('income_type')['total_income'].median()
# –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω—ã–π –¥–æ—Ö–æ–¥, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–∏–ø–µ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏
for i in group_income.index:
    data.loc[(data['income_type'] == i) & (data['total_income'].isnull()), 'total_income'] = group_income[i]
    
#–ø—Ä–æ–≤–µ—Ä–∏–º –∑–∞–º–µ–Ω—É
print('–£—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ NA', data['total_income'].isna().sum())
print()
#–ø—Ä–æ–ø—É—Å–∫ –ø–æ–ª–∞ –≤ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ
print(data[data['gender'] == 'XNA'])
print()
#—É–¥–∞–ª–∏–º –µ–≥–æ
data = data.drop(data[data['gender'] == 'XNA'].index)
#–ø—Ä–æ–≤–µ—Ä–∏–º
print(data['gender'].value_counts())
#–≤—Å–µ –æ–∫

# –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤–æ–∑—Ä–∞—Å—Ç–∞ –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–∏–ø–µ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ 
#–¢–∞–∫ –Ω–∞–ø—Ä–∏–º–µ—Ä –º–æ–∂–Ω–æ –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –≤–æ–∑—Ä–∞—Å—Ç –ø–µ–Ω—Å–∏–æ–Ω–µ—Ä–æ–≤, –Ω–∞–ø—Ä–∏–º–µ—Ä.
group_dop_years = data.groupby('income_type')['dob_years'].median()
# –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–∏–ø–µ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏
for i in group_dop_years.index:
    data.loc[(data['income_type'] == i) & (data['dob_years'] == 0), 'dob_years'] = group_dop_years[i]

#–æ–±—Ä–∞–±–æ—Ç–∞–µ–º —Å—Ç–∞–∂
group_days_employed_income = data.groupby('income_type')['days_employed'].mean()
print(group_days_employed_income)
#–¥—É–º–∞—é –º–æ–∂–Ω–æ —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–π —Å—Ç–∞–∂ –ø–æ–¥–µ–ª–∏—Ç—å –Ω–∞ 24, –ø–æ—Ç–æ–º –Ω–∞ 365, —Ç–∞–∫ –∫–∞–∫ –ø–æ—Ö–æ–∂–µ –Ω–∞ —Ç–æ, —á—Ç–æ –æ–Ω –ø–æ–ø—Ä–æ—Å—Ç—É –∏–∑–º–µ—Ä—è–ª—Å—è –≤ —á–∞—Å–∞—Ö. –î–ª—è 
#–∞–¥–µ–∫–≤–∞—Ç–Ω—ã—Ö —Ç–∏–ø–æ–≤ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ –º—ã –ø—Ä–æ—Å—Ç–æ –¥–µ–ª–∏–º –Ω–∞ 365. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å–æ–∑–¥–∞–µ—Ç—Å—è –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ —Å—Ç–∞–∂–∞ –≤ –≥–æ–¥–∞—Ö.
#–∞ —Ç–∞–∫–∂–µ –∏–∑–±–∞–≤–∏–º—Å—è –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ñ—É–Ω–∫—Ü–∏–∏ abs
data.loc[data['income_type'].str.lower().str.contains('–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π|–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä'), 'days_employed'] = abs(data['days_employed'] / 365 /24)
data.loc[data['income_type'].str.lower().str.contains('–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π|–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä') == False, 'days_employed'] = abs(data['days_employed'] / 365)

group_days_employed_income = data.groupby('income_type')['days_employed'].mean()
print(group_days_employed_income)

# –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ days_employed –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–∏–ø–µ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏
group_days_employed = data.groupby('income_type')['days_employed'].median()
# –∑–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω—ã–π –≤–æ–∑—Ä–∞—Å—Ç, –æ—Å–Ω–æ–≤—ã–≤–∞—è—Å—å –Ω–∞ —Ç–∏–ø–µ –∑–∞–Ω—è—Ç–æ—Å—Ç–∏
for i in group_days_employed.index:
    data.loc[(data['income_type'] == i) & (data['days_employed'].isnull()), 'days_employed'] = group_days_employed[i]
```

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

```python
#–Ω–∞–º –Ω—É–∂–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ —Ç–∞–±–ª–∏—Ü–µ
def purpose_category(purpose):
    purpose_l = m.lemmatize(purpose)
    if ('–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å' in purpose_l) or ('–∂–∏–ª—å–µ' in purpose_l):
        prefix = '–ø—Ä–∏–æ–±—Ä–µ—Ç–µ–Ω–∏–µ '
        #if '—Ä–µ–º–æ–Ω—Ç' in purpose_l:
        #    prefix = '—Ä–µ–º–æ–Ω—Ç '
        #elif '—Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ' in purpose_l:
        #    prefix = '—Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ '
        #else:
        #    prefix = '–ø—Ä–∏–æ–±—Ä–µ—Ç–µ–Ω–∏–µ '
        #
        #if '–∂–∏–ª–æ–π' in purpose_l:
        #    return prefix+'–∂–∏–ª–æ–π –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
        #elif '—Å–≤–æ–π' in purpose_l:
        #    return prefix+'–∂–∏–ª–æ–π –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
        #elif '–∫–æ–º–º–µ—Ä—á–µ—Å–∫–∏–π' in purpose_l:
        #    return prefix+'–∫–æ–º–º–µ—Ä—á–µ—Å–∫–æ–π –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
        #elif '—Å–¥–∞—á–∞' in purpose_l:
        #    return prefix+'–∫–æ–º–º–µ—Ä—á–µ—Å–∫–æ–π –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
        #else:
        return prefix+'–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
    elif '–∞–≤—Ç–æ–º–æ–±–∏–ª—å' in purpose_l:
        return '–∞–≤—Ç–æ–º–æ–±–∏–ª—å'
    elif '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ' in purpose_l:
        return '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'
    elif '—Å–≤–∞–¥—å–±–∞' in purpose_l:
        return '—Å–≤–∞–¥—å–±–∞'
    
def income_category(income:int, p25:int, p50:int, p75:int):
    if (income <= p25):
        return '–Ω–∏–∑–∫–∏–π –¥–æ—Ö–æ–¥'
    elif (p25 < income) and (income <= p50):
        return '—Å—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥'
    elif (p50 < income) and (income <= p75):
        return '–≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ'
    else:
        return '–≤—ã—Å–æ–∫–∏–π –¥–æ—Ö–æ–¥'

#–¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ–∏—Ä–µ–π —Ü–µ–ª–∏
data['purpose_category'] = data['purpose'].apply(purpose_category)
#
print(data['purpose_category'].value_counts())
#–í—ã–≤–µ—Å—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Å –ø–æ–º–æ—â—å—é describe
data_descr = data.describe()
data_descr['total_income'] = data_descr['total_income'].map(lambda x: '{:.0f}'.format(x))
display(data_descr)
#print(data_descr.info())
perc25 = int(data_descr.loc['25%', 'total_income'])
perc50 = int(data_descr.loc['50%', 'total_income'])
perc75 = int(data_descr.loc['75%', 'total_income'])
#–∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∏—Ä—É–µ–º —É—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ –≤ –≥—Ä—É–ø–ø—ã –ø–æ –ø–µ—Ä—Ü–µ–ø—Ç–∏–ª—è–º
#display(data.info())
data['total_income_category'] = data['total_income'].apply(income_category, args=(perc25, perc50, perc75))
print()
print(data['total_income_category'].value_counts())
```

* –ö–∞–∫ –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —Å—Ç–∞—Ç—É—Å –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—Ç–Ω–æ—à–µ–Ω–∏–π –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ —Å–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –Ω–∞ –≤–æ–∑–≤—Ä–∞—Ç–µ –∫—Ä–µ–¥–∏—Ç–∞. –õ—é–¥–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Å–æ—Å—Ç–æ—è—Ç –≤ –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω—ã—Ö –æ—Ç–Ω–æ—à–µ–Ω–∏—è—Ö –∏ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ —Å–æ—Å—Ç–æ—è–ª–∏ —Ö—É–∂–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∫—Ä–µ–¥–∏—Ç—ã, –≤ –æ—Ç–ª–∏—á–∏–∏ –æ—Ç —Å–µ–º–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω—ã–π —Å—Ç–∞—Ç—É—Å, –ø—É—Å—Ç—å –¥–∞–∂–µ –∫–æ–≥–¥–∞-—Ç–æ –≤ –ø—Ä–æ—à–ª–æ–º.
    
P.s. —Ç—ã –∑–Ω–∞–ª, —á—Ç–æ –º–æ–∂–Ω–æ —á—É—Ç—å –±–æ–ª–µ–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ —Å—Ç—Ä–æ–∏—Ç—å —Å–≤–æ–¥–Ω—ã–µ —Ç–∞–±–ª–∏—á–∫–∏? –°–º–æ—Ç—Ä–∏:

```python
# –ö–û–î –†–ï–í–¨–Æ–í–ï–†–ê

import plotly.express as px

def pivot(data, index):
    
    """
    –Ø —Ä–µ—à–∏–ª –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∏—Ç —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—á–∫—É –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ—Ç –µ–µ, —Ç—É—Ç –Ω–µ—Ç –Ω–∏—á–µ–≥–æ —Å–ª–æ–∂–Ω–æ–≥–æ, —Å–º–æ—Ç—Ä–∏:
    –Ω–∞ –≤—Ö–æ–¥ –æ–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∏ –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏, –¥–∞–ª–µ–µ —Å—Ç—Ä–æ–∏—Ç —Å–≤–æ–¥–Ω—É—é —Å –ø–æ–¥—Å—á–µ—Ç–æ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ (count),
    —á–∏—Å–ª–∞ –∑–∞–µ–º—â–∏–∫–æ–≤ (sum), % –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞ (mean). –¢–∫ –ø—Ä–∏–∑–Ω–∞–∫ debt –±–∏–Ω–∞—Ä–Ω—ã–π, —Ç–æ —Ç–∞–∫–∏–µ –∞–≥–≥—Ä–µ–≥–∏—Ä—É—é—â–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∫–∞–∫ –Ω–µ–ª—å–∑—è –∫—Å—Ç–∞—Ç–∏.
    
    
    count - –ø–æ–¥—Å—á–∏—Ç–∞–µ—Ç —á–∏—Å—Ç–æ –≤—Å–µ —Å—Ç—Ä–æ–∫–∏, —Ç.–µ.  –æ–±—â–µ–µ —á–∏—Å–ª–æ –∫–ª–∏–µ–Ω—Ç–æ–≤
    sum - —Å—É–º–º—É –Ω–æ–ª–∏–∫–æ–≤ –∏ –µ–¥–∏–Ω–∏—á–µ–∫, —Ç.–µ. –ø–æ —Ñ–∞–∫—Ç—É –±—É–¥–µ—Ç —Ç–æ–ª—å–∫–æ —Å—É–º–º–∞ –µ–¥–∏–Ω–∏—Ü => –∫–æ–ª-–≤–æ –¥–æ–ª–∂–Ω–∏–∫–æ–≤
    mean - —Å—É–º–º–∞ –¥–µ–ª–µ–Ω–Ω–∞ –Ω–∞ –∫–æ–ª-–≤–æ, —Ç.–µ. —Ç–µ –∂–µ –¥–æ–ª–∂–Ω–∏–∫–∏ –¥–µ–ª–µ–Ω–Ω—ã–µ –Ω–∞ –æ–±—â–µ–µ —á–∏—Å–ª–æ –∫–ª–∏–µ–Ω—Ç–æ–≤
    
    –ö–ª–∞—Å—Å–Ω–æ –∂–µ!)
    
    """
    
    pivot = data.pivot_table(index = index, values = 'debt', aggfunc = ['count', 'sum', 'mean']).reset_index()
    pivot.columns = [index, '–ö–æ–ª-–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤', '–ö–æ–ª-–≤–æ –¥–æ–ª–∂–Ω–∏–∫–æ–≤', '% –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞']
    display(pivot.sort_values('% –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞', ascending = False))
    
    fig = px.bar(pivot, x = index, y = '% –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞', color = '% –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞', title = '% –Ω–µ–≤–æ–∑–≤—Ä–∞—Ç–∞ –ø–æ ' + index)
    fig.show()
    

pivot(data, 'family_status')
```
