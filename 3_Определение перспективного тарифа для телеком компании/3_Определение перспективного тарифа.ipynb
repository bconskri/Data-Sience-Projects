## Описание проекта

Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.  
Предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов.  
В распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

### Описание тарифов
#### Тариф «Смарт»
Ежемесячная плата: 550 рублей  
Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика  
Стоимость услуг сверх тарифного пакета:  
минута разговора: 3 рубля  
сообщение: 3 рубля  
1 Гб интернет-трафика: 200 рублей  

#### Тариф «Ультра»
Ежемесячная плата: 1950 рублей  
Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика  
Стоимость услуг сверх тарифного пакета:  
минута разговора: 1 рубль  
сообщение: 1 рубль  
1 Гб интернет-трафика: 150 рублей  

«Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута.

### Посчитаем обьемы ресурсов по месяцам для каждого пользователя

Количество сделанных звонков и израсходованных минут разговора по месяцам;  
количество отправленных сообщений по месяцам;  
объем израсходованного интернет-трафика по месяцам;  
помесячную выручку с каждого пользователя (вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану).

```python
#создание сводной таблицы - количества ресурсов по месяцам
def by_month(data, val_column, func):
    return data.pivot_table(index = ['user_id', 'month'], values = val_column, aggfunc = func)\
               .reset_index(inplace = False).fillna(0)

#создадим массивы с количествами израсходованных ресурсов по месяцам по всем пользователям в том числе и с кол-вом звонков
calls_by_month = by_month(calls, 'duration', 'count')
minutes_by_month = by_month(calls, 'duration', 'sum')
messages_by_month = by_month(messages, 'id', 'count')
mbs_by_month = by_month(internet, 'mb_used', 'sum')

#обьеденим таблицы
aggregate_data = (calls_by_month.merge(minutes_by_month, on=['user_id', 'month'], how='outer')
                                .merge(messages_by_month, on=['user_id', 'month'], how='outer')
                                .merge(mbs_by_month, on=['user_id', 'month'], how='outer').fillna(0))
aggregate_data.columns = ['user_id', 'month', 'calls', 'duration', 'messages', 'mb_used']

#приклеим тариф
aggregate_data = aggregate_data.merge(users[['user_id', 'tariff_id']], on='user_id', how='left')

display(aggregate_data)
```

Так у нас каждая строка будет содержать в себе данные по одному месяцу и пользователю, после этого мы можем объединить наши таблицы через 2 столбца, используя метод `outer` - поскольку у нас могут быть пользователи, которые могут пользоваться одной из услуг, но не пользоваться остальными.
```python
#присоединим данные тарифов
aggregate_data = aggregate_data.merge(tariffs[['minutes_included','rub_per_minute','messages_included',\
                                               'rub_per_message', 'mb_per_month_included','rub_per_gb', 'rub_monthly_fee']], \
                                      left_on='tariff_id', right_index=True, how='left').fillna(0)

over_limit_data = copy.deepcopy(aggregate_data)

display(over_limit_data)

#функция расчёта превышения
def over_limit(row, column, limit_column):
    result = int(row[column]) - int(row[limit_column])
    if result > 0:
        return result
    elif result <= 0:
        return 0

#рассчитаем построчно перелимит для всех ресурсов и переведем в рублb
over_limit_data['duration_overlimit_rub'] = over_limit_data.apply(over_limit, axis = 1, \
                                            args = ['duration', 'minutes_included'])*over_limit_data['rub_per_minute'] 
over_limit_data['messages_overlimit_rub'] = over_limit_data.apply(over_limit, axis = 1, \
                                            args = ['messages', 'messages_included'])*over_limit_data['rub_per_message'] 
over_limit_data['mb_used_overlimit_rub'] = over_limit_data.apply(over_limit, axis = 1, \
                                           args = ['mb_used', 'mb_per_month_included'])*over_limit_data['rub_per_gb']/1024

display(over_limit_data)
```

```python
#теперь получим помесячную выручку с каждого пользователя
column_to_drop = ['month', 'minutes_included', 'rub_per_minute', 'messages_included', 'rub_per_message', \
                  'mb_per_month_included', 'rub_per_gb']
profit_per_user = over_limit_data.groupby(by=['user_id', 'tariff_id']).sum().reset_index().drop(columns=column_to_drop)
#сложим все перелимиты и размер абон платы
profit_per_user['total_income'] = profit_per_user['rub_monthly_fee'] + profit_per_user['duration_overlimit_rub'] + \
                            profit_per_user['messages_overlimit_rub'] + profit_per_user['mb_used_overlimit_rub']

display(profit_per_user)
```

## 4. Проверим гипотезы

- средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

### Гипотеза 1. Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается
Нулевой гипотезой `H0` будет гипотеза о том, что средние наших выборок равны. Т.е. выручки не различаются.  
Альтернативной гипотезой `H1` будет гипотеза "Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается".

Используем метод `scipy.stats.ttest_ind (array1, array2, equal_var = False)`. `False`, потому что дисперсии наших выборок отличаются.

Критический уровень значимости установим 0.05. Хочу чтобы вероятность правильности вывода была 95% и больше.

```python
alpha = 0.05
results = st.ttest_ind(profit_per_user[profit_per_user['tariff_id'] == 0]['total_income'], \
                       profit_per_user[profit_per_user['tariff_id'] == 1]['total_income'], equal_var = False)
print('P-value: {:.30f}'.format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Вывод по гипотезе 1:

Нулевой гипотезой была гипотеза о равенстве средних наших выборок и ничтожно маленькое p-value говорит нам о том, вероятность случайного совпадения средних наших выборок ничтожно мало. Нулевая гипотеза отвергается.  
А значит мы принимаем альтернативную гипотезу о том, что средние наших выборок отличаются. Т.е. Средние выручки пользователей разных тарифов - разные.

### Гипотеза 2. Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов
